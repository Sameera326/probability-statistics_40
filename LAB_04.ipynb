```python
# Analysis Questions
# 1. Most common price range
# Answer printed above after histogram analysis

# 2. Correlation between Horsepower and Mileage
# Look at the scatterplot and trendline: There seems to be a negative correlation between horsepower and mileage (powerful cars tend to have lower mileage).

# 3. Most popular car brand
# Answer printed above after bar chart visualization

# 4. Fuel type preference trends over the years
# Based on the grouped bar chart, it looks like electric cars have become more popular in recent years, especially in the 2019-2022 range.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('car_data_250.csv')
```

```python
# 1. Histogram Analysis - Price distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Price Distribution of Cars')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Identify the most common price range (bin with highest frequency)
price_range = pd.cut(df['Price'], bins=20).value_counts().idxmax()
print(f"Most common price range: {price_range}")
```

```python
# 2. Scatterplot Exploration - Horsepower vs. Mileage
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Horsepower', y='Mileage', hue='Fuel Type', palette='deep')
plt.title('Horsepower vs. Mileage')
plt.xlabel('Horsepower')
plt.ylabel('Mileage (mpg)')
plt.grid(True)

# Adding a trendline (Linear fit)
sns.regplot(data=df, x='Horsepower', y='Mileage', scatter=False, color='black', line_kws={"linestyle": "--"})
plt.show()
```

```python
# 3. Bar Chart Visualization - Number of cars by Brand
plt.figure(figsize=(10, 6))
brand_counts = df['Brand'].value_counts()
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette='viridis')
plt.title('Number of Cars by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()

# Identify the most popular brand based on number of cars
most_popular_brand = brand_counts.idxmax()
print(f"The most popular brand is: {most_popular_brand}")
```

```python
# 4. Grouped Bar Chart - Number of Cars by Fuel Type within Manufacturing Years
plt.figure(figsize=(12, 8))
fuel_year_counts = pd.crosstab(df['Year'], df['Fuel Type'])
fuel_year_counts.plot(kind='bar', stacked=True, figsize=(12, 8), cmap='Set2')
plt.title('Number of Cars by Fuel Type Across Manufacturing Years')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.legend(title='Fuel Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
