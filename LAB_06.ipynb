```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
mean = 60
std_dev = 12
num_samples = 1000
data = np.random.normal(mean, std_dev, num_samples)
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, stat='density', color='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean = 30
std_dev = 5

# Probabilities
p_greater_equal_19 = 1 - stats.norm.cdf(19, mean, std_dev)
p_less_equal_19 = stats.norm.cdf(19, mean, std_dev)
p_between_19_35 = stats.norm.cdf(35, mean, std_dev) - stats.norm.cdf(19, mean, std_dev)

# Print the probabilities
print(f"Probability of daily calorie intake >= 19 calories: {p_greater_equal_19:.4f}")
print(f"Probability of daily calorie intake <= 19 calories: {p_less_equal_19:.4f}")
print(f"Probability of daily calorie intake between 19 and 35 calories: {p_between_19_35:.4f}")

# Visualization
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = stats.norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='black')
plt.fill_between(x, y, where=(x >= 19), alpha=0.3, color='red', label='>= 19 calories')
plt.fill_between(x, y, where=(x <= 19), alpha=0.3, color='blue', label='<= 19 calories')
plt.fill_between(x, y, where=((x >= 19) & (x <= 35)), alpha=0.3, color='green', label='19 to 35 calories')

plt.title('Probability Visualization of Daily Calorie Intake')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Density')
plt.legend()
plt.show()
```
