```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
mean = 60
std_dev = 12
num_samples = 1000
data = np.random.normal(mean, std_dev, num_samples)
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, stat='density', color='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define the PDF of a normal distribution
def normal_pdf(x, mean, std_dev):
    var = std_dev ** 2
    denom = np.sqrt(2 * np.pi * var)
    num = np.exp(-(x - mean) ** 2 / (2 * var))
    return num / denom
# Define the CDF of a normal distribution using a numerical approximation
def normal_cdf(x, mean, std_dev, num_points=1000):
    # Create an array of points from -10 to x
    xs = np.linspace(-10 * std_dev + mean, x, num_points)
    # Calculate the PDF at each point
    pdf_vals = normal_pdf(xs, mean, std_dev)
    # Approximate the CDF by integrating the PDF
    cdf = np.trapz(pdf_vals, xs)
    return cdf
# Set the parameters for the normal distribution
mean_calories = 30
std_dev_calories = 5
# Calculate probabilities
P_greater_equal_19 = 1 - normal_cdf(19, mean_calories, std_dev_calories)
P_less_equal_19 = normal_cdf(19, mean_calories, std_dev_calories)
P_between_19_and_35 = normal_cdf(35, mean_calories, std_dev_calories) - normal_cdf(19, mean_calories, std_dev_calories)
# Print the probabilities
print("Probability of daily calorie intake greater than or equal to 19 calories:", P_greater_equal_19)
print("Probability of daily calorie intake less than or equal to 19 calories:", P_less_equal_19)
print("Probability of daily calorie intake between 19 and 35 calories:", P_between_19_and_35)
# Visualization
x = np.linspace(10, 50, 1000)
pdf = normal_pdf(x, mean_calories, std_dev_calories)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', label='PDF of Daily Calorie Intake')
plt.fill_between(x, pdf, where=(x >= 19), color='lightblue', alpha=0.5, label='P(X >= 19)')
plt.fill_between(x, pdf, where=(x <= 19), color='lightgreen', alpha=0.5, label='P(X <= 19)')
plt.fill_between(x, pdf, where=(x >= 19) & (x <= 35), color='black', alpha=0.5, label='P(19 <= X <= 35)')
# Add labels and title
plt.title('Probability Density Function of Daily Calorie Intake')
plt.xlabel('Calories')
plt.ylabel('Density')
plt.legend()
# Show the plot
plt.show()
```
