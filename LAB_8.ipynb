```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Data
calories = [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000]
exercise_duration = [60, 50, 30, 90, 45, 20, 75, 95, 15, 10]
bmi = [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]

# Scatter Plots
plt.figure(figsize=(15, 5))
```

```python
# Daily Calorie Intake vs. BMI
plt.subplot(1, 3, 1)
plt.scatter(calories, bmi, color='blue')
plt.title('Daily Calorie Intake vs. BMI')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('BMI')
```

```python
# Exercise Duration vs. BMI
plt.subplot(1, 3, 2)
plt.scatter(exercise_duration, bmi, color='green')
plt.title('Exercise Duration vs. BMI')
plt.xlabel('Exercise Duration (minutes)')
plt.ylabel('BMI')
```

```python
# Daily Calorie Intake vs. Exercise Duration
plt.subplot(1, 3, 3)
plt.scatter(calories, exercise_duration, color='red')
plt.title('Daily Calorie Intake vs. Exercise Duration')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Exercise Duration (minutes)')

plt.tight_layout()
plt.show()
```

```python
# Pearson Correlation Coefficients
corr_cal_bmi, _ = pearsonr(calories, bmi)
corr_ex_bmi, _ = pearsonr(exercise_duration, bmi)
corr_cal_ex, _ = pearsonr(calories, exercise_duration)
```

```python
# Display Results
print(f"Pearson Correlation Coefficient (Calories vs. BMI): {corr_cal_bmi:.2f}")
print(f"Pearson Correlation Coefficient (Exercise vs. BMI): {corr_ex_bmi:.2f}")
print(f"Pearson Correlation Coefficient (Calories vs. Exercise): {corr_cal_ex:.2f}")
```
